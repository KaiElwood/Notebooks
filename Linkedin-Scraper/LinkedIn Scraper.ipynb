```python
#Get page source code
src = browser.page_source
soup = BeautifulSoup(src, 'lxml')
#Strip text from source code
results = soup.find('small', {'class': 'display-flex t-12 t-black--light t-normal'}).get_text().strip().split()[0]
results = int(results.replace(',', ''))
print(results)
```

```python
# jobs-search-results__list-item
# div > job-card-container--clickable
jobResults = pd.DataFrame()
for job in browser.find_elements_by_class_name('jobs-search-results__list-item'):
    companyName = job.find_element_by_class_name('job-card-container__company-name').text
    jobHref = job.find_element_by_class_name('job-card-list__title').get_attribute('href')
    jobTitle = job.find_element_by_class_name('job-card-list__title').text
    
    jobInfo = {'company': companyName, 'link': jobHref, 'title': jobTitle}
    jobResults.append(jobInfo, ignore_index=True)
```
