```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# functional
plt.plot(x,y, 'r-')
plt.xlabel('X Label')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")
plt.title("Y over X")

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.title("X over Y")
```

```python
# OO Method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig1 = plt.figure()

axes1 = fig1.add_axes([.1,.1,.8,.8])
axes2 = fig1.add_axes([0.2,.4,.4,.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(ncols=2,nrows=2,figsize=(16,6))

# this sets the figure to be a certain width and height, and then attaches some data to the figures via the axes
# it will not automatically expand to accomodate more the the ncols*nrows setup
# but you can use all of the ncols and nrows accordingly (see below)

axes[0][0].plot(x,y)

axes[1][0].plot(y,x)
plt.tight_layout
```
